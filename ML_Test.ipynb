```python
import pandas as pd
import numpy  as np
```

```python
data={"Apple":[1,2,4,8,16,32,64,128,256,512,1024, 2048, 4096],
      "Price":[10,15,400,60,None,280, 640,1280,None,41200,4240,20480,3096]}
```

```python
df=pd.DataFrame(data)
df
```

```python
for col in df:
  if str(df[col].dtype) !="object":
    q1= df[col].quantile(q=0.25)
    q3= df[col].quantile(q=0.75)
    iqr= q3-q1
    lower_bound= q1-1.5*iqr
    upper_bound= q3+1.5*iqr
    print(col, np.where(df[col]<lower_bound))
    print(col, np.where(df[col]>upper_bound))


    df.drop(np.where(df[col]<lower_bound)[0], inplace=True)
    df.drop(np.where(df[col]>upper_bound)[0], inplace= True)
    print(col, np.where(df[col]<lower_bound))
    print(col, np.where(df[col]>upper_bound))
```
