Find the price of 13" & 15" pizza
  
      

```python
import pandas as pd
```

```python
data={"Size":[10,12,14,16],
      "Price":[5.99,7.99,9.99,11.99]}
df=pd.DataFrame(data)
df
```

```python
def meann(lst):
  total_count =0
  for i in lst:
    total_count=total_count+i
  mean=total_count/len(lst)
  return mean

```

```python
xb, yb= meann(df["Size"]), meann(df["Price"])
```
